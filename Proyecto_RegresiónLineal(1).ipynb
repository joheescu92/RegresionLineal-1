```python
# Importar librerías
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import pandas as pd
import os
```

```python
los_datos= pd.read_csv('/content/drive/My Drive/RegresionLineal/Ejemplo.csv', sep=";")
```

```python
los_datos
```

```python
#Defino las variables
X = los_datos["experiencia"].values.reshape(-1, 1)
y = los_datos["salario"].values
# Entreno el modelo con los datos de experiencia y salario
model = LinearRegression()
model.fit(X, y)
```

```python
# Predicciones
X_test = np.array([[0], [12]])  # Usamos el mismo rango de los datos originales
y_pred = model.predict(X_test)
```

```python
# Visualizo los resultados
plt.scatter(X, y, color='blue', label='Datos')
plt.plot(X_test, y_pred, color='red', linewidth=2, label='Pronosticos')
plt.xlabel('Años de Experiencia (X)')
plt.ylabel('Salario Anual (y)')
plt.legend()
plt.title('Ejemplo de Regresión Lineal')
plt.grid(True)
plt.show()
```

```python
# Imprimo los coeficientes del modelo
intercepto = model.intercept_
coeficiente = model.coef_[0]
(intercepto, coeficiente)
```

```python
# Calculo el coeficiente de determinación (R^2) usando el método score
r2_score = model.score(X, y)
```

```python
# Muestro el resultado
print(f"Coeficiente de determinación (R^2): {r2_score:.4f}")
```
